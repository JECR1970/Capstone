```python
#'date' to datetime
df_calendar['date']=pd.to_datetime(df_calendar['date'])
```

```python
#we create a new column 'week_of_the_day'
df_calendar['week_of_the_day'] = df_calendar['date'].dt.isocalendar().week
```

```python
df_calendar.head()
```

```python
df_calendar_copy=df_calendar.copy()
```

```python
df_calendar['event'].fillna(0, inplace=True)
```

```python
df_calendar.info()
```

```python
#this for loop to check if every weekday matches always with its weekday_int (ex: every 'Saturday' matches with a '1', and so on...).
dayweek = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
j=1
for i in dayweek:
  print(df_calendar[(df_calendar['weekday']==i) & (df_calendar['weekday_int']!=j)])
  j=j+1
```

```python
#Since the 'weekday' and the 'weekday_int' are perfect matches, we can get rid of the 'weekday' column
df_calendar1=df_calendar
del(df_calendar['weekday'])
df_calendar1.head()
```

```python
df_calendar.info()
```

```python
#This for loop to delete the prefix 'd_' from 'd' values...
a=0
for i in df_calendar['d']:
  df_calendar['d'][a]=df_calendar['d'][a].replace('d_', '')
  a+=1
```
