##Import item_prices.csv

```python
df_prices = pd.read_csv('/content/drive/MyDrive/Capstonedrive/item_prices.csv')
```

```python
df_prices.tail()
```

```python
df_prices.info()
```

```python
#only ['yearweek'] has nulls, and the %nulls is 3.5%
df_prices['yearweek'].isnull().value_counts()/len(df_prices)*100
```

```python
#we drop the null rows
df_prices.drop(df_prices[df_prices['yearweek'].isnull()].index, inplace=True)
```

```python
#and check that there are no nulls left
df_prices['yearweek'].isnull().value_counts()/len(df_prices)*100
```

```python
df_prices.reset_index(inplace=True)
```

```python
del(df_prices['index'])
```

```python
df_prices.info()
```

```python
df_prices['item'].value_counts()
```

```python
#['item'] to numeric using LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df_prices['itemEncoded'] = le.fit_transform(df_prices['item'])
```

```python
df_prices = df_prices.reindex(['item', 'itemEncoded', 'category', 'store_code', 'yearweek', 'sell_price'], axis=1)
```

```python
df_prices.head()
```

```python
df_prices.info()
```

```python
#this calculates the number of categories and the number of items within each of them
df_prices['category'].value_counts()
```

```python
#['category'] to numeric, using map()
dcat=df_prices['category'].map({'ACCESORIES':0, 'HOME_&_GARDEN': 1, 'SUPERMARKET': 2})
```

```python
#this calculates the number of stores and the number of different items carried by each one of them, in ascending order.
df_prices['store_code'].value_counts()
```

```python
#['store_code'] to numeric, using map()
dstore=df_prices['store_code'].map({'BOS_1': 11, 'BOS_2': 12, 'BOS_3': 13, 'NYC_1': 21, 'NYC_2': 22, 'NYC_3': 23, 'NYC-4': 24, 'PHI_1': 31, 'PHI_2': 32, 'PHI_3': 33})
```

```python
df_prices.rename(columns={'category':'cat', 'store_code':'store'}, inplace=True)
```

```python
df_prices=pd.concat([df_prices, dstore, dcat], axis=1)
```

```python
df_prices.head()
```

```python
df_prices.drop(['cat', 'store'], axis=1, inplace=True)
```

```python
df_prices.head()
```

```python
df_prices = df_prices.reindex(['item', 'itemEncoded', 'category', 'store_code', 'yearweek', 'sell_price'], axis=1)
```

```python
df_prices.info()
```

```python
df_prices.tail()
```

##Import daily_calendar_with_events.csv

```python
#'date' to datetime
df_calendar['date']=pd.to_datetime(df_calendar['date'])
```

```python
#we create a new column 'week_of_the_day'
df_calendar['week_of_the_day'] = df_calendar['date'].dt.isocalendar().week
```

```python
df_calendar.head()
```

```python
df_calendar_copy=df_calendar.copy()
```

```python
df_calendar.isnull().value_counts()
```

```python
df_calendar[df_calendar['event']!=0]
```

```python
df_calendar['event'].fillna(0, inplace=True)
```

```python
df_calendar.info()
```

```python
#this for loop to check if every weekday matches always with its weekday_int (ex: every 'Saturday' matches with a '1', and so on...).
dayweek = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
j=1
for i in dayweek:
  print(df_calendar[(df_calendar['weekday']==i) & (df_calendar['weekday_int']!=j)])
  j=j+1
```

```python
#Since the 'weekday' and the 'weekday_int' are perfect matches, we can get rid of the 'weekday' column
df_calendar1=df_calendar
del(df_calendar['weekday'])
df_calendar1.head()
```

```python
df_calendar.info()
```

```python
#This for loop to delete the prefix 'd_' from 'd' values...
a=0
for i in df_calendar['d']:
  df_calendar['d'][a]=df_calendar['d'][a].replace('d_', '')
  a+=1
```
